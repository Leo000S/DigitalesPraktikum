## Einfaches Diagramm mit Matplotlib

Jetzt machen wir ein **Liniendiagramm** mit der Bibliothek `matplotlib`.

> Hinweis: In JupyterLite ist matplotlib bereits vorinstalliert.

```python
import matplotlib.pyplot as plt

# Daten
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Plot
plt.plot(x, y, label="y = x²", color='green')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ein einfaches Liniendiagramm")
plt.legend()
plt.grid(True)
plt.show()
```
